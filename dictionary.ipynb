```python
car_dict={"brand":"BMW","model":"ix","year":1964}
print(car_dict)
```

#dictionaries

```python
#accesing dictionary elements
print(car_dict["brand"])
```

```python
#length of dictionary
print(len(car_dict))
```

```python
#types of data
print(type(car_dict))
```

```python
# for duplicate keys items
car_dict2={"brand":"BMW","model":"ix","year":1964,"year":2000}
print(car_dict2)
```

```python
# for accessing dict items
x=car_dict.get('model')
print(x)
```

```python
# the key() method will return a list of all the keys in the dictionary.
car_dict2.keys()
```

```python
# to change values of dictionary
car_dict2["year"]=2024
car_dict2.values()
```

```python
car_dict2["color"]="white"
car_dict2.values()
```

```python
#the values in dictionary items can be of any data type
d3={"name":"A",'year':2023,'pass':"true",'marks':[50,60,70]}
print(d3)
```

# update method

```python
d3.update({'year':2024})
print(d3)
```

```python
d3['phone_no']=94566
print(d3)
```

```python
d3.update({'email':'asdf@gamil.com'})
print(d3)
```

```python
# for removing dictionry items
# the pop() method removes the item with the specified key name :

d3.pop('phone_no')
```
