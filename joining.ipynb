#joining sets


```python
#union method
set1={"x","y","z"}
set2={6,7,8}
set3=set1.union(set2)
print(set3)
```

```python
#opreator
set1={"x","y","z"}
set2={6,7,8}
set3=set1|set2
print(set3)
```

```python
#join multiple sets
set1={"x","y","z"}
set2={6,7,8}
set3={"motu","patlu"}
set4={"king","kong"}
largeset=set1.union(set2,set3,set4)
print(largeset)
```

```python
set1={"x","y","z"}
set2={6,7,8}
set3={"motu","patlu"}
set4={"king","kong"}
largeset=set1 | set2|set3|set4
print(largeset)
```

```python
#updaet
set1={"x","y","z"}
set2={6,7,8}
set1.update(set2)
print(set1)
```

```python
set1={"x","y","z"}
set2={1,2,3,4,5,'x'}
set1=set1.intersection(set2)
print(set1)
```

```python
# & operator
set1={"x","y","z"}
set2={6,7,8,"x"}
set1=set1 & set2
print(set1)
```

```python
# difference method
set1={"x","y","z"}
set2={6,7,8,"x"}
set3=set1.difference(set2)
print(set3)
```
